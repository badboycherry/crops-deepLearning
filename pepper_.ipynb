```python
import torch
from ultralytics import YOLO
```

```python
torch.cuda.is_available()
```

```python
model = YOLO('yolov8n-cls.pt')
```

```python
if torch.cuda.is_available():
  model.to('cuda')
```

```python
results = model.train(data='/content/drive/MyDrive/pepper_data', epochs=10, imgsz=640, save=True, project='/content/drive/MyDrive', name='/content/drive/MyDrive/pepper_train')
```

```python
# 'ClassifyMetrics' 객체에서 결과 딕셔너리 가져오기
results_dict = results.results_dict

# 클래스 이름과 확률을 출력
for key, value in results_dict.items():
    if 'top1' in key:  # top-1 클래스의 경우
        top1_class = key.split('_')[-1]  # 클래스 이름 추출
        top1_probability = value
        print(f'Top-1 Class: {top1_class}')
        print(f'Top-1 Probability: {top1_probability}')
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from ultralytics import YOLO  # YOLO 모델 불러오기
import os
import random
import matplotlib.patches as patches
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"  # 나눔고딕 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

rows = 1  # 행의 수를 1으로 설정
cols = 3  # 열의 수를 3로 설정
# 모델 및 클래스 레이블 설정
trained_model = YOLO('/content/drive/MyDrive/pepper_train/weights/best.pt')
class_labels = {0: 'pepper_healthy', 1: 'pepper_mildew', 2: 'pepper_tan'}

# 이미지 폴더 경로 설정
folder_path = '/content/drive/MyDrive/1조 공유/abv'

# 폴더 내의 모든 이미지 파일 경로 가져오기
image_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

# 랜덤으로 하나의 이미지 선택
random_image_path = random.choice(image_files)

fig, axes = plt.subplots(rows, cols, figsize=(20, 15))  # 1행 3열의 subplot 설정
axes = axes.flatten()  # axes 배열을 1차원으로 변환

for ax, image_path in zip(axes, random.sample(image_files, rows * cols)):  # 3개의 이미지 랜덤 선택
    results = trained_model(image_path)
    img = Image.open(image_path)
    ax.imshow(img)

    # 결과에 따른 타이틀 및 사각형 설정
    top1_class_index = results[0].probs.top1
    top1_confidence = results[0].probs.top1conf.item()
    predicted_label = class_labels[top1_class_index]

    if top1_class_index == 10:  # 라벨 지정
        rect_color = 'green'
        result_text = '성공'
    else:  # 실패
        rect_color = 'red'
        result_text = '실패'

    rect = patches.Rectangle((0, 0), img.width-1, img.height-1, linewidth=4, edgecolor=rect_color, facecolor='none')
    ax.add_patch(rect)
    ax.set_title(f"예측된 클래스: {predicted_label}, 확률: {top1_confidence:.2f}\n결과: {result_text}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
